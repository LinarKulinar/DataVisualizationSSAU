### Что делает geom_col()? Чем он отличается от geom_bar()?

```python
# Type your code here...
```

`geom_col()` представляет значения данных в виде столбцов.

`geom_bar()` делает высоту полосы пропорциональной количеству наблюдений в каждой группе.

то есть `geom_col()` - это альтернативная версия geom_bar, которая выставляет высоту столбцов в зависимости от какой-либо переменной в данных.

```python
ggplot(data=diamonds) +\
geom_col(mapping=aes(x="cut", y="clarity"))
```

### Зачем тут coord_fixed()? Что делает geom_abline()?

`coord_fixed()` - задает соотношение сторон для системы координат графика

`geom_abline()` строит диагональную линию с заданным наклоном и сдвигом

```python
ggplot(data=mpg, mapping=aes(x="cty", y="hwy")) +\
geom_point() +\
geom_abline() +\
coord_fixed()
```
